```python
# Importing the 'networkx1'

import networkx as nx
```

```python
# Need matplotlib library to draw the graph
import matplotlib.pyplot as plt
```

## Creating Graph

```python
# Creating an empty graph
newG = nx.Graph()
```

```python
# Accessing the vertex set and the edge set of the graph newG

# Returns a list (which is empty now)
print(newG.nodes())

# Returns a list (which is empty now)
print(newG.edges())
```

```python
# Adding single node
newG.add_node("A")
```

```python
# Adding multiple nodes at once
newG.add_nodes_from(["B", "C", "D", "E"])
```

```python
# Adding an edge connecting two nodes A and B
newG.add_edge(*("A", "B"))  # single edge as tuple of two nodes
```

Let's see the Graph so far

```python
nx.draw(newG)
plt.show()
```

```python
# Adding multiple edges
newG.add_edges_from([("A", "C"), ("B", "D"), ("B", "E"), ("C", "E")])
```

Now, accessing the sets of all the Vertices and edges

```python
# Getting the Vertices
print("Vertex set: ", newG.nodes())
```

```python
# Getting the Edges
print("Edge set: ", newG.edges())
```

```python
nx.draw(newG)
plt.show()
```

```python
# We can save the generated figure in any desired file format
# plt.savefig("graph.png") # or eps, jpeg, etc.
```

```python
# Using the attributes of nx.draw() to draw the vertex labels
nx.draw(newG, with_labels=True, font_weight='bold')
plt.show()
```

Creating Peterson's graph

```python
GP = nx.petersen_graph()
```

```python
nx.draw(GP, with_labels=True, font_weight='bold')
plt.show()
```

Adjacency View


```python
# Getting the adjacency list
print(newG.adj)
```

Degree of a vertex

```python
print(newG.degree("A"))
```

## Creating Weighted Graph

```python
elementList = [('A', 'B', 2), 
               ('A', 'C', 1),
               ('B', 'D', 5),
               ('B', 'E', 3),
               ('C', 'E', 2)]

weightedG = nx.Graph()

weightedG.add_weighted_edges_from(elementList)
```

```python
# Drawing a weighted graph

pos = nx.spring_layout(weightedG)

nx.draw(weightedG, pos, with_labels=True, font_weight='bold')

edge_weight = nx.get_edge_attributes(weightedG, 'weight')

nx.draw_networkx_edge_labels(weightedG, pos, edge_labels=edge_weight)

plt.show()
```
